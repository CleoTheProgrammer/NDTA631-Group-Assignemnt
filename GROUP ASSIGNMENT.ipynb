```python
ict_sa = ict[ict["REF_AREA"] == "ZAF"].copy()
cyb_sa = cyb[cyb["REF_AREA"] == "ZAF"].copy()
print("\nICT South Africa shape:", ict_sa.shape)
print("Cyber South Africa shape:", cyb_sa.shape)
```

```python
# Detect year columns
def find_year_columns(df: pd.DataFrame):
    years = []
    for col in df.columns:
        if isinstance(col, int) and 2014 <= col <= 2025:
            years.append(col)
        elif isinstance(col, str) and len(col) == 4 and col.isdigit():
            years.append(col)
    
    return sorted(years, key=lambda x: int(str(x)))
    
    
ict_years = find_year_columns(ict_sa)
cyb_years = find_year_columns(cyb_sa)
```

```python
print(ict_sa.describe())
```

```python
print(cyb_sa.describe())
```

```python
# Reshape wide → long
def to_long(df, years):
    long_df = df.melt(
        id_vars=["REF_AREA","REF_AREA_LABEL","INDICATOR","INDICATOR_LABEL"],
        value_vars=years,
        var_name="Year",
        value_name="Value"
    )
    long_df["Year"] = pd.to_numeric(long_df["Year"], errors="coerce")
    long_df["Value"] = pd.to_numeric(long_df["Value"], errors="coerce")
    return long_df

ict_long = to_long(ict_sa, ict_years)
cyb_long = to_long(cyb_sa, cyb_years)
print("\nICT Long format sample:")
print(ict_long.head(5))
print("\nCyber Long format sample:")
print(cyb_long.head(5))
```

```python
# Pivot to Year × Indicator
ict_pivot = ict_long.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")
cyb_pivot = cyb_long.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")
print("\nICT Pivot (Year × Indicators):", ict_pivot.shape)
print(ict_pivot.head(3))
print("\nCyber Pivot (Year × Indicators):", cyb_pivot.shape)
print(cyb_pivot.head(3))
```

```python
# YoY analysis
arr = ict_pivot.to_numpy(dtype=float)
ict_diff = np.diff(arr, axis=0)
ict_pct  = np.divide(ict_diff, arr[:-1], out=np.full_like(ict_diff, np.nan), where=arr[:-1]!=0)
print("\nICT YoY Absolute Change sample:")
print(pd.DataFrame(ict_diff, index=ict_pivot.index[1:], columns=ict_pivot.columns).head(3))
print("\nICT YoY % Change sample:")
print(pd.DataFrame(ict_pct, index=ict_pivot.index[1:], columns=ict_pivot.columns).head(3))
```

```python
# Composite indices
ict_comp = np.nanmean(arr, axis=1)
print("\nICT Composite sample (first 5 years):")
print(dict(zip(ict_pivot.index[:5], ict_comp[:5])))
```

```python
# CAGR per indicator
def cagr_per_indicator(df_long):
    rows = []
    for ind, g in df_long.sort_values("Year").groupby("INDICATOR_LABEL"):
        g = g.dropna(subset=["Value"])
        if g.empty: continue
        first_val, last_val = g.iloc[0]["Value"], g.iloc[-1]["Value"]
        n_years = int(g.iloc[-1]["Year"] - g.iloc[0]["Year"])
        if n_years > 0 and first_val not in [0, np.nan]:
            cagr = (last_val / first_val) ** (1/n_years) - 1
            rows.append((ind, cagr))
    return pd.DataFrame(rows, columns=["Indicator","CAGR"]).sort_values("CAGR", ascending=False)

ict_cagr = cagr_per_indicator(ict_long)
print("\nICT CAGR sample:")
print(ict_cagr.head(5))
```

```python
TOP_POS = 8          # top positive movers to display
TOP_NEG = 8          # top negative movers to display
LABEL_MAX = 48       # max characters for indicator labels

df = ict_pivot.copy()

# Compute YoY % change
yoy_abs = df.diff()
with np.errstate(divide='ignore', invalid='ignore'):
    yoy_pct = (yoy_abs / df.shift(1)) * 100.0
yoy_pct = yoy_pct.replace([np.inf, -np.inf], np.nan)

# Select the most recent year with any YoY data
if yoy_pct.dropna(how='all').empty:
    raise ValueError("No YoY data available to plot.")
last_yoy = yoy_pct.dropna(how='all').iloc[-1].dropna()

# Split into top risers and fallers
sorted_yoy = last_yoy.sort_values()
fallers = sorted_yoy.head(TOP_NEG)
risers  = sorted_yoy.tail(TOP_POS)
sel = pd.concat([fallers, risers])

# Prepare labels and values
labels = [shorten(str(k), width=LABEL_MAX, placeholder="…") for k in sel.index]
values = sel.values
ypos = np.arange(len(sel))

# Plot horizontal bars for readability
fig_h = max(6, 0.45 * len(sel))
fig, ax = plt.subplots(figsize=(12, fig_h))
ax.barh(ypos, values)
ax.set_yticks(ypos)
ax.set_yticklabels(labels, fontsize=9)
ax.set_xlabel("YoY % change (most recent year)")
ax.set_title("ICT — Latest-Year YoY % (Top Risers & Fallers)")

# Zero reference line
ax.axvline(0, linestyle='--', linewidth=1)

# Annotate values
for y, v in zip(ypos, values):
    ax.text(v + (0.5 if v >= 0 else -0.5), y, f"{v:.1f}%", va='center',
            ha='left' if v >= 0 else 'right', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
TOP_N = 12          # number of indicators to display
LABEL_MAX = 42      # max characters for labels

df = ict_pivot.copy()

# Keep years with any data; find global start/end years with broad coverage
valid_years = df.dropna(how='all').index
start_year = int(valid_years.min())
end_year   = int(valid_years.max())

# Rank indicators: prioritize completeness across the full window, then variance
comp_full = df.loc[[start_year, end_year]].notna().sum(axis=0)  # need data in both years
variance  = pd.Series(np.nanvar(df.to_numpy(float), axis=0), index=df.columns)

cands = pd.DataFrame({"comp_full": comp_full, "var": variance}).dropna()
ranked = cands.sort_values(["comp_full", "var"], ascending=[False, False]).index.tolist()

# Select top-N indicators that have both start and end values
cols = []
for c in ranked:
    if len(cols) >= TOP_N: break
    if pd.notna(df.at[start_year, c]) and pd.notna(df.at[end_year, c]):
        cols.append(c)

if not cols:
    raise ValueError("No indicators have values at both the start and end year.")

# Build start/end series
start = df.loc[start_year, cols]
end   = df.loc[end_year, cols]

# Sort by absolute change for a clean story
order = (end - start).abs().sort_values(ascending=True).index  # small→large slope for readability
start = start[order]
end   = end[order]

# Plot
y = np.arange(len(order))
fig_h = max(6, 0.45 * len(order))
fig, ax = plt.subplots(figsize=(12, fig_h))

for i, ind in enumerate(order):
    ax.plot([0, 1], [start[i], end[i]], marker='o')
    ax.text(-0.02, start[i], f"{start[i]:.2f}", va='center', ha='right', fontsize=8)
    ax.text( 1.02, end[i],   f"{end[i]:.2f}",   va='center', ha='left',  fontsize=8)

# Y-axis labeling with indicator names in the middle (between start and end)
mid_y = (start.values + end.values) / 2
for i, ind in enumerate(order):
    ax.text(0.5, mid_y[i], shorten(str(ind), width=LABEL_MAX, placeholder="…"),
            ha='center', va='center', fontsize=9)

ax.set_xlim(-0.15, 1.15)
ax.set_xticks([0, 1])
ax.set_xticklabels([str(start_year), str(end_year)])
ax.set_title(f"ICT — Start vs End Levels (Slopegraph, {start_year} → {end_year})")
ax.set_ylabel("Level")
ax.grid(axis='y', alpha=0.2)

plt.tight_layout()
plt.show()
```

```python
# --- Helper: z-score normalisation within each indicator (column) ---
def zscore_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    # compute mean/std per column ignoring NaNs
    col_mean = df.mean(axis=0, skipna=True)
    col_std  = df.std(axis=0, skipna=True)
    # avoid division by zero
    col_std = col_std.replace(0, np.nan)
    return (df - col_mean) / col_std

# Normalise each pillar so indicators are comparable before averaging
ict_norm = zscore_df(ict_pivot)
cyb_norm = zscore_df(cyb_pivot)

# Composite = NaN-aware mean across indicators per year
ict_comp = ict_norm.mean(axis=1, skipna=True)
cyb_comp = cyb_norm.mean(axis=1, skipna=True)

# Align to common South Africa years where at least one pillar has a value
years_common = sorted(set(ict_comp.dropna().index).intersection(set(cyb_comp.dropna().index)))
if not years_common:
    raise ValueError("No common years with composite values in both ICT and Cyber.")

ict_line = ict_comp.loc[years_common]
cyb_line = cyb_comp.loc[years_common]

# Plot
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ict_line.index.astype(int), ict_line.values, marker='o', label="ICT (normalized composite)")
ax.plot(cyb_line.index.astype(int), cyb_line.values, marker='o', label="Cyber (normalized composite)")

ax.set_title("ICT vs Cyber — Normalized Composite (Common Years, South Africa)")
ax.set_xlabel("Year")
ax.set_ylabel("Composite (z-score mean)")
ax.legend()
plt.tight_layout()
plt.show()
```

> ## Insights

>>On the common-year window, the normalized composites place ICT and Cyber on the same scale (z-scores): values above 0 indicate performance above each pillar’s long-run average and values below 0 indicate below average. The ICT line sits consistently above 0 and above the Cyber line, signalling relative outperformance and sustained positive momentum. Cyber starts below its average and rises into positive territory by the end of the window, indicating a catch-up from a lower base. The spread between the two pillars narrows, but ICT maintains the lead. Importantly, these indicators reflect system capability and maturity, not public awareness—so the data support a maturity gap (ICT adoption/capacity ahead of Cyber capability) rather than any claim about what people “knew”.
>>
> ## Conclusion
>>
>> On a like-for-like, normalized basis, ICT outperforms Cyber across the period, maintaining an above-average trajectory while Cyber improves from below-average to above-average, narrowing (but not closing) the gap. This pattern suggests South Africa has scaled access, capacity, and usage faster than it has strengthened cybersecurity readiness, i.e., adoption is outpacing protection.ICT’s structural position remains stronger; Cyber shows meaningful progress from a lower starting point.

```python
KEYWORDS = ["price", "basket", "voice", "data-only", "usage", "bundle", "tariff", "5gb"]
TOP_N = 6            # max number of price indicators to plot
LABEL_MAX = 50       # max characters for legend labels

df = ict_long.copy()

# Ensure numeric
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["Value"] = pd.to_numeric(df["Value"], errors="coerce")

# Filter to likely price/affordability indicators by label keywords (case-insensitive)
mask = df["INDICATOR_LABEL"].str.lower().fillna("").apply(
    lambda s: any(k in s for k in KEYWORDS)
)
price_df = df[mask].copy()

if price_df.empty:
    raise ValueError("No pricing/affordability indicators found. Adjust KEYWORDS to match your labels.")

# Pivot to Year × Indicator for processing
price_pivot = price_df.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")

# Rank indicators by completeness, then variance (most informative first)
completeness = price_pivot.notna().sum(axis=0)
variance = pd.Series(np.nanvar(price_pivot.to_numpy(float), axis=0), index=price_pivot.columns)
ranked = (pd.DataFrame({"comp": completeness, "var": variance})
          .sort_values(["comp", "var"], ascending=[False, False])).index.tolist()

# Keep top-N
cols = ranked[:TOP_N]
price_pivot = price_pivot[cols]

# Determine common years across selected indicators and choose base year
valid_years_per_col = [set(price_pivot[c].dropna().index.astype(int)) for c in price_pivot.columns]
common_years = sorted(set.intersection(*valid_years_per_col)) if valid_years_per_col else []
if len(common_years) < 2:
    raise ValueError("Not enough common years across selected price indicators to index. Try increasing TOP_N or adjusting KEYWORDS.")
base_year = common_years[0]

# Build indexed series (Base Year = 100) and drop any series with invalid base values
indexed = {}
for c in price_pivot.columns:
    s = price_pivot[c].dropna()
    if base_year not in s.index or s.loc[base_year] in [0, np.nan]:
        continue
    indexed[c] = (s.loc[common_years] / s.loc[base_year]) * 100.0

if not indexed:
    raise ValueError("No valid series could be indexed (base year missing or zero).")

# Plot
fig, ax = plt.subplots(figsize=(12, 5))
for label, series in indexed.items():
    ax.plot(series.index.astype(int), series.values, marker='o', label=shorten(str(label), width=LABEL_MAX, placeholder="…"))

ax.set_title(f"ICT Prices/Affordability — Indexed to {base_year} = 100 (South Africa)")
ax.set_xlabel("Year")
ax.set_ylabel("Index (Base Year = 100)")

# Reference line at 100 (base)
ax.axhline(100, linestyle='--', linewidth=1)

# Place legend outside for readability
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load the datasets
ict = pd.read_csv("data/gci_ICT.csv")
cyb = pd.read_csv("data/gci_Cybersecurity.csv")

print("ICT DataFrame shape:", ict.shape)
print("Cybersecurity DataFrame shape:", cyb.shape)

# Filter to South Africa only
ict_sa = ict[ict["REF_AREA"] == "ZAF"].copy()
cyb_sa = cyb[cyb["REF_AREA"] == "ZAF"].copy()

print("\nSouth Africa ICT data shape:", ict_sa.shape)
print("South Africa Cybersecurity data shape:", cyb_sa.shape)

# Identify year columns (assuming they are 4-digit strings)
def find_year_columns(df):
    years = []
    for col in df.columns:
        if isinstance(col, str) and len(col) == 4 and col.isdigit():
            years.append(col)
    return sorted(years, key=lambda x: int(x))

ict_years = find_year_columns(ict_sa)
cyb_years = find_year_columns(cyb_sa)

print("\nICT year columns:", ict_years)
print("Cybersecurity year columns:", cyb_years)

# Reshape data from wide to long format
def melt_dataframe(df, year_columns, id_columns=None):
    if id_columns is None:
        id_columns = ["REF_AREA", "REF_AREA_LABEL", "INDICATOR", "INDICATOR_LABEL"]
    
    melted_df = df.melt(
        id_vars=id_columns,
        value_vars=year_columns,
        var_name="Year",
        value_name="Value"
    )
    
    # Convert to appropriate data types
    melted_df["Year"] = pd.to_numeric(melted_df["Year"], errors="coerce")
    melted_df["Value"] = pd.to_numeric(melted_df["Value"], errors="coerce")
    
    return melted_df

ict_long = melt_dataframe(ict_sa, ict_years)
cyb_long = melt_dataframe(cyb_sa, cyb_years)

print("\nICT long format shape:", ict_long.shape)
print("Cybersecurity long format shape:", cyb_long.shape)

# Handle missing values using forward and backward fill within each indicator group
def handle_missing_values(df):
    df["Value"] = df.groupby("INDICATOR_LABEL")["Value"].transform(
        lambda x: x.ffill().bfill()
    )
    return df

ict_long = handle_missing_values(ict_long)
cyb_long = handle_missing_values(cyb_long)

# Verify no missing values remain
print("\nMissing values in ICT data:", ict_long["Value"].isna().sum())
print("Missing values in Cybersecurity data:", cyb_long["Value"].isna().sum())

# Display sample of cleaned data
print("\nSample of cleaned ICT data:")
print(ict_long.head())
print("\nSample of cleaned Cybersecurity data:")
print(cyb_long.head())
```

```python
# Step 2: Conditional Formatting and Analysis
print("="*60)
print("STEP 2: CONDITIONAL FORMATTING AND ANALYSIS")
print("="*60)

# Define key indicators for analysis
key_ict_indicators = [
    "Active mobile-broadband subscriptions per 100 inhabitants (ITU)",
    "Fixed-broadband subscriptions per 100 inhabitants (ITU)",
    "Percentage of individuals using the internet (ITU)",
    "Mobile-cellular subscriptions per 100 inhabitants (ITU)",
    "International bandwidth usage - per capita",
    "Percentage of households with internet access at home (ITU)"
]

key_cyb_indicators = [
    "Global Cybersecurity Index - Overall Score (ITU GCI)",
    "Capacity Development Score - CDS (ITU GCI)",
    "Cooperative Score - CS (ITU GCI)",
    "Technical Score - TS (ITU GCI)",
    "Organizational Score - OS (ITU GCI)",
    "Cybercrime Regulation - LS2 (ITU GCI)",
    "Child Online Protection - TS4 (ITU GCI)",
    "Training for Cybersecurity Professionals - CDS2 (ITU GCI)"
]

# Filter data for key indicators
ict_key = ict_long[ict_long["INDICATOR_LABEL"].isin(key_ict_indicators)].copy()
cyb_key = cyb_long[cyb_long["INDICATOR_LABEL"].isin(key_cyb_indicators)].copy()

print(f"Filtered {len(ict_key)} records for ICT analysis")
print(f"Filtered {len(cyb_key)} records for Cybersecurity analysis")

# Create pivot tables for easier analysis
ict_pivot = ict_key.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")
cyb_pivot = cyb_key.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")

print("\nICT Pivot Table Shape:", ict_pivot.shape)
print("Cybersecurity Pivot Table Shape:", cyb_pivot.shape)

# Define conditional formatting functions
def apply_conditional_formatting(df, high_threshold=None, low_threshold=None):
    """
    Apply conditional formatting to identify high and low values
    """
    formatted_df = df.copy()
    
    if high_threshold is not None and low_threshold is not None:
        # For ICT indicators with custom thresholds
        formatted_df["Status"] = np.where(
            formatted_df["Value"] > high_threshold, 
            "High", 
            np.where(formatted_df["Value"] < low_threshold, "Low", "Normal")
        )
    else:
        # For cybersecurity scores (0-100 scale)
        formatted_df["Status"] = np.where(
            formatted_df["Value"] >= 80, 
            "Excellent", 
            np.where(formatted_df["Value"] >= 60, "Good", 
                    np.where(formatted_df["Value"] >= 40, "Fair", "Poor"))
        )
    
    return formatted_df

# Apply conditional formatting to ICT data
print("\nApplying conditional formatting to ICT data...")
ict_conditional_list = []

for indicator in key_ict_indicators:
    indicator_data = ict_key[ict_key["INDICATOR_LABEL"] == indicator].copy()
    if not indicator_data.empty and len(indicator_data) > 1:
        # Calculate thresholds based on data distribution
        q1 = indicator_data["Value"].quantile(0.25)
        q3 = indicator_data["Value"].quantile(0.75)
        
        formatted_data = apply_conditional_formatting(
            indicator_data, 
            high_threshold=q3,
            low_threshold=q1
        )
        ict_conditional_list.append(formatted_data)

# Combine all ICT conditional data
if ict_conditional_list:
    ict_conditional = pd.concat(ict_conditional_list, ignore_index=True)
else:
    ict_conditional = pd.DataFrame()
    print("Warning: No ICT data available for conditional formatting")

print(f"ICT conditional formatting applied to {len(ict_conditional)} records")

# Apply conditional formatting to Cybersecurity data
print("Applying conditional formatting to Cybersecurity data...")
cyb_conditional_list = []

for indicator in key_cyb_indicators:
    indicator_data = cyb_key[cyb_key["INDICATOR_LABEL"] == indicator].copy()
    if not indicator_data.empty:
        formatted_data = apply_conditional_formatting(indicator_data)
        cyb_conditional_list.append(formatted_data)

# Combine all Cybersecurity conditional data
if cyb_conditional_list:
    cyb_conditional = pd.concat(cyb_conditional_list, ignore_index=True)
else:
    cyb_conditional = pd.DataFrame()
    print("Warning: No Cybersecurity data available for conditional formatting")

print(f"Cybersecurity conditional formatting applied to {len(cyb_conditional)} records")

# Display samples of formatted data
if not ict_conditional.empty:
    print("\nSample of ICT Data with Conditional Formatting:")
    internet_indicator = "Percentage of individuals using the internet (ITU)"
    if internet_indicator in ict_conditional["INDICATOR_LABEL"].values:
        ict_sample = ict_conditional[ict_conditional["INDICATOR_LABEL"] == internet_indicator]
        print(ict_sample[["Year", "Value", "Status"]].head(10))
    else:
        print(f"Indicator '{internet_indicator}' not found in ICT data")
        print("Available indicators:", ict_conditional["INDICATOR_LABEL"].unique())
else:
    print("No ICT data available for display")

if not cyb_conditional.empty:
    print("\nSample of Cybersecurity Data with Conditional Formatting:")
    gci_indicator = "Global Cybersecurity Index - Overall Score (ITU GCI)"
    if gci_indicator in cyb_conditional["INDICATOR_LABEL"].values:
        cyb_sample = cyb_conditional[cyb_conditional["INDICATOR_LABEL"] == gci_indicator]
        print(cyb_sample[["Year", "Value", "Status"]].head())
    else:
        print(f"Indicator '{gci_indicator}' not found in Cybersecurity data")
        print("Available indicators:", cyb_conditional["INDICATOR_LABEL"].unique())
else:
    print("No Cybersecurity data available for display")

# Calculate year-over-year changes
def calculate_yoy_changes(df):
    df = df.sort_values(["INDICATOR_LABEL", "Year"])
    df["YoY_Change"] = df.groupby("INDICATOR_LABEL")["Value"].pct_change() * 100
    df["YoY_Absolute_Change"] = df.groupby("INDICATOR_LABEL")["Value"].diff()
    return df

print("\nCalculating year-over-year changes...")
ict_analysis = calculate_yoy_changes(ict_key)
cyb_analysis = calculate_yoy_changes(cyb_key)

print("Sample of ICT Data with YoY Analysis:")
internet_analysis = ict_analysis[ict_analysis["INDICATOR_LABEL"] == "Percentage of individuals using the internet (ITU)"]
print(internet_analysis[["Year", "Value", "YoY_Change", "YoY_Absolute_Change"]].head(10))

print("\nSample of Cybersecurity Data with YoY Analysis:")
gci_analysis = cyb_analysis[cyb_analysis["INDICATOR_LABEL"] == "Global Cybersecurity Index - Overall Score (ITU GCI)"]
print(gci_analysis[["Year", "Value", "YoY_Change", "YoY_Absolute_Change"]].head())

# Identify significant changes (greater than 10% or 5 points)
significant_changes_ict = ict_analysis[
    (abs(ict_analysis["YoY_Change"]) > 10) | (abs(ict_analysis["YoY_Absolute_Change"]) > 5)
].dropna(subset=["YoY_Change"])

significant_changes_cyb = cyb_analysis[
    (abs(cyb_analysis["YoY_Change"]) > 10) | (abs(cyb_analysis["YoY_Absolute_Change"]) > 5)
].dropna(subset=["YoY_Change"])

print(f"\nSignificant ICT changes (>10% or >5 points): {len(significant_changes_ict)} records")
print(f"Significant Cybersecurity changes (>10% or >5 points): {len(significant_changes_cyb)} records")

# Display most significant changes
if not significant_changes_ict.empty:
    print("\nTop 5 Significant ICT Changes:")
    top_ict_changes = significant_changes_ict.nlargest(5, "YoY_Change")[["INDICATOR_LABEL", "Year", "Value", "YoY_Change"]]
    print(top_ict_changes)

if not significant_changes_cyb.empty:
    print("\nTop 5 Significant Cybersecurity Changes:")
    top_cyb_changes = significant_changes_cyb.nlargest(5, "YoY_Change")[["INDICATOR_LABEL", "Year", "Value", "YoY_Change"]]
    print(top_cyb_changes)

# Create summary statistics for conditional formatting
if not ict_conditional.empty:
    ict_summary = ict_conditional.groupby(["INDICATOR_LABEL", "Status"]).size().unstack(fill_value=0)
    print("\nICT Conditional Formatting Summary:")
    print(ict_summary)
else:
    print("\nNo ICT data available for conditional formatting summary")

if not cyb_conditional.empty:
    cyb_summary = cyb_conditional.groupby(["INDICATOR_LABEL", "Status"]).size().unstack(fill_value=0)
    print("\nCybersecurity Conditional Formatting Summary:")
    print(cyb_summary)
else:
    print("\nNo Cybersecurity data available for conditional formatting summary")

# Calculate overall statistics
ict_stats = ict_key.groupby("INDICATOR_LABEL")["Value"].agg(["mean", "std", "min", "max"]).round(2)
cyb_stats = cyb_key.groupby("INDICATOR_LABEL")["Value"].agg(["mean", "std", "min", "max"]).round(2)

print("\nICT Value Statistics:")
print(ict_stats)

print("\nCybersecurity Value Statistics:")
print(cyb_stats)

# Identify years with the highest and lowest values for each indicator
def get_extreme_years(df):
    result = []
    for indicator in df["INDICATOR_LABEL"].unique():
        indicator_data = df[df["INDICATOR_LABEL"] == indicator]
        if not indicator_data.empty:
            max_idx = indicator_data["Value"].idxmax()
            min_idx = indicator_data["Value"].idxmin()
            result.append({
                "Indicator": indicator,
                "Max_Year": indicator_data.loc[max_idx, "Year"],
                "Max_Value": indicator_data.loc[max_idx, "Value"],
                "Min_Year": indicator_data.loc[min_idx, "Year"],
                "Min_Value": indicator_data.loc[min_idx, "Value"]
            })
    return pd.DataFrame(result)

ict_extremes = get_extreme_years(ict_key)
cyb_extremes = get_extreme_years(cyb_key)

print("\nICT Extreme Values:")
print(ict_extremes)

print("\nCybersecurity Extreme Values:")
print(cyb_extremes)

print("\nStep 2 completed successfully! Conditional formatting and analysis applied.")
```

```python
# Step 5: Export Final Analysis
print("="*60)
print("STEP 5: EXPORT FINAL ANALYSIS")
print("="*60)

import os
import json

# Create analysis_results directory if it doesn't exist
os.makedirs("analysis_results", exist_ok=True)

# Export cleaned data and analysis results
print("Exporting cleaned data...")
ict_long.to_csv("analysis_results/cleaned_ict_south_africa.csv", index=False)
cyb_long.to_csv("analysis_results/cleaned_cybersecurity_south_africa.csv", index=False)

ict_analysis.to_csv("analysis_results/ict_analysis_with_yoy.csv", index=False)
cyb_analysis.to_csv("analysis_results/cybersecurity_analysis_with_yoy.csv", index=False)

# Export conditional formatting results if they exist
if not ict_conditional.empty:
    ict_conditional.to_csv("analysis_results/ict_conditional_formatting.csv", index=False)
if not cyb_conditional.empty:
    cyb_conditional.to_csv("analysis_results/cybersecurity_conditional_formatting.csv", index=False)

# Create summary statistics with proper error handling
print("Creating summary statistics...")
summary_stats = {
    "ICT_Indicators_Count": len(ict_key["INDICATOR_LABEL"].unique()),
    "Cybersecurity_Indicators_Count": len(cyb_key["INDICATOR_LABEL"].unique()),
    "Years_Analyzed": f"{ict_analysis['Year'].min()}-{ict_analysis['Year'].max()}",
}

# Calculate average internet growth rate safely
internet_growth_data = ict_analysis[ict_analysis['INDICATOR_LABEL'] == 'Percentage of individuals using the internet (ITU)']
if not internet_growth_data.empty and 'YoY_Change' in internet_growth_data.columns:
    avg_growth = internet_growth_data['YoY_Change'].mean()
    summary_stats["Average_Internet_Growth_Rate"] = f"{avg_growth:.2f}%"
else:
    summary_stats["Average_Internet_Growth_Rate"] = "N/A"

# Calculate cybersecurity score improvement safely
cybersecurity_score_name = "Global Cybersecurity Index - Overall Score (ITU GCI)"
if not cyb_pivot.empty and cybersecurity_score_name in cyb_pivot.columns:
    available_years = cyb_pivot.index.tolist()
    if len(available_years) >= 2:
        first_year, last_year = available_years[0], available_years[-1]
        improvement = cyb_pivot.loc[last_year, cybersecurity_score_name] - cyb_pivot.loc[first_year, cybersecurity_score_name]
        summary_stats["Cybersecurity_Score_Improvement"] = f"{improvement:.2f} points ({first_year}-{last_year})"
    else:
        summary_stats["Cybersecurity_Score_Improvement"] = "Insufficient years for comparison"
else:
    summary_stats["Cybersecurity_Score_Improvement"] = "N/A"

# Add additional statistics
if not ict_stats.empty:
    summary_stats["Average_ICT_Values"] = ict_stats["mean"].to_dict()
if not cyb_stats.empty:
    summary_stats["Average_Cybersecurity_Values"] = cyb_stats["mean"].to_dict()

# Count significant changes
summary_stats["Significant_ICT_Changes_Count"] = len(significant_changes_ict)
summary_stats["Significant_Cybersecurity_Changes_Count"] = len(significant_changes_cyb)

# Add data quality metrics
summary_stats["ICT_Data_Completeness"] = f"{(1 - ict_long['Value'].isna().mean()) * 100:.1f}%"
summary_stats["Cybersecurity_Data_Completeness"] = f"{(1 - cyb_long['Value'].isna().mean()) * 100:.1f}%"

# Save summary statistics
with open("analysis_results/summary_statistics.json", "w") as f:
    json.dump(summary_stats, f, indent=4)

print("Summary statistics saved to 'analysis_results/summary_statistics.json'")

# Create a detailed report CSV
print("Creating detailed report...")
detailed_report = []

# ICT indicators report
for indicator in ict_key["INDICATOR_LABEL"].unique():
    indicator_data = ict_key[ict_key["INDICATOR_LABEL"] == indicator]
    if not indicator_data.empty:
        report_entry = {
            "Category": "ICT",
            "Indicator": indicator,
            "First_Year": indicator_data["Year"].min(),
            "Last_Year": indicator_data["Year"].max(),
            "Min_Value": indicator_data["Value"].min(),
            "Max_Value": indicator_data["Value"].max(),
            "Average_Value": indicator_data["Value"].mean(),
            "Data_Points": len(indicator_data)
        }
        detailed_report.append(report_entry)

# Cybersecurity indicators report
for indicator in cyb_key["INDICATOR_LABEL"].unique():
    indicator_data = cyb_key[cyb_key["INDICATOR_LABEL"] == indicator]
    if not indicator_data.empty:
        report_entry = {
            "Category": "Cybersecurity",
            "Indicator": indicator,
            "First_Year": indicator_data["Year"].min(),
            "Last_Year": indicator_data["Year"].max(),
            "Min_Value": indicator_data["Value"].min(),
            "Max_Value": indicator_data["Value"].max(),
            "Average_Value": indicator_data["Value"].mean(),
            "Data_Points": len(indicator_data)
        }
        detailed_report.append(report_entry)

# Convert to DataFrame and save
detailed_report_df = pd.DataFrame(detailed_report)
detailed_report_df.to_csv("analysis_results/detailed_indicators_report.csv", index=False)
print("Detailed report saved to 'analysis_results/detailed_indicators_report.csv'")

# Create a trends analysis report
print("Creating trends analysis report...")
trends_report = []

# Analyze trends for each ICT indicator
for indicator in ict_key["INDICATOR_LABEL"].unique():
    indicator_data = ict_analysis[ict_analysis["INDICATOR_LABEL"] == indicator].dropna(subset=["YoY_Change"])
    if not indicator_data.empty and len(indicator_data) > 1:
        avg_growth = indicator_data["YoY_Change"].mean()
        max_growth = indicator_data["YoY_Change"].max()
        min_growth = indicator_data["YoY_Change"].min()
        
        trends_report.append({
            "Category": "ICT",
            "Indicator": indicator,
            "Average_Growth_Rate": f"{avg_growth:.2f}%",
            "Max_Growth_Rate": f"{max_growth:.2f}%",
            "Min_Growth_Rate": f"{min_growth:.2f}%",
            "Trend": "Increasing" if avg_growth > 0 else "Decreasing" if avg_growth < 0 else "Stable"
        })

# Analyze trends for each Cybersecurity indicator
for indicator in cyb_key["INDICATOR_LABEL"].unique():
    indicator_data = cyb_analysis[cyb_analysis["INDICATOR_LABEL"] == indicator].dropna(subset=["YoY_Change"])
    if not indicator_data.empty and len(indicator_data) > 1:
        avg_growth = indicator_data["YoY_Change"].mean()
        max_growth = indicator_data["YoY_Change"].max()
        min_growth = indicator_data["YoY_Change"].min()
        
        trends_report.append({
            "Category": "Cybersecurity",
            "Indicator": indicator,
            "Average_Growth_Rate": f"{avg_growth:.2f}%",
            "Max_Growth_Rate": f"{max_growth:.2f}%",
            "Min_Growth_Rate": f"{min_growth:.2f}%",
            "Trend": "Increasing" if avg_growth > 0 else "Decreasing" if avg_growth < 0 else "Stable"
        })

# Convert to DataFrame and save
trends_report_df = pd.DataFrame(trends_report)
trends_report_df.to_csv("analysis_results/trends_analysis_report.csv", index=False)
print("Trends analysis saved to 'analysis_results/trends_analysis_report.csv'")

# Create a file manifest
file_manifest = {
    "files_created": [
        "cleaned_ict_south_africa.csv",
        "cleaned_cybersecurity_south_africa.csv",
        "ict_analysis_with_yoy.csv",
        "cybersecurity_analysis_with_yoy.csv",
        "ict_conditional_formatting.csv",
        "cybersecurity_conditional_formatting.csv",
        "summary_statistics.json",
        "detailed_indicators_report.csv",
        "trends_analysis_report.csv",
        "south_africa_ict_cybersecurity_dashboard.png",
        "yoy_analysis.png",
        "significant_ict_changes.png",
        "significant_cybersecurity_changes.png"
    ],
    "analysis_date": pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S"),
    "data_sources": ["gci_ICT.csv", "gci_Cybersecurity.csv"],
    "analysis_period": f"{ict_analysis['Year'].min()}-{ict_analysis['Year'].max()}"
}

with open("analysis_results/file_manifest.json", "w") as f:
    json.dump(file_manifest, f, indent=4)

print("File manifest saved to 'analysis_results/file_manifest.json'")

print("\n" + "="*60)
print("ANALYSIS COMPLETED SUCCESSFULLY!")
print("="*60)
print("\nSummary Statistics:")
for key, value in summary_stats.items():
    print(f"{key}: {value}")

print(f"\nFiles created in 'analysis_results' folder:")
for file in file_manifest["files_created"]:
    print(f"  - {file}")

print(f"\nAnalysis completed on: {file_manifest['analysis_date']}")
print(f"Data period analyzed: {file_manifest['analysis_period']}")
```
