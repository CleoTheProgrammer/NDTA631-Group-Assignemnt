```python

print("ICT DataFrame loaded:", ict.shape)
print(ict.head(3))
print("Cyber DataFrame loaded:", cyb.shape)
print(cyb.head(3))
```

```python
# Filter to South Africa
ict_sa = ict[ict["REF_AREA"] == "ZAF"].copy()
cyb_sa = cyb[cyb["REF_AREA"] == "ZAF"].copy()
print("\nICT South Africa shape:", ict_sa.shape)
print("Cyber South Africa shape:", cyb_sa.shape)
```

```python
# Detect year columns
def find_year_columns(df: pd.DataFrame):
    years = []
    for col in df.columns:
        if isinstance(col, int) and 2014 <= col <= 2025:
            years.append(col)
        elif isinstance(col, str) and len(col) == 4 and col.isdigit():
            years.append(col)
    
    return sorted(years, key=lambda x: int(str(x)))
    
    
ict_years = find_year_columns(ict_sa)
cyb_years = find_year_columns(cyb_sa)


print(ict_sa.describe())
```

```python
# Reshape wide → long
def to_long(df, years):
    long_df = df.melt(
        id_vars=["REF_AREA","REF_AREA_LABEL","INDICATOR","INDICATOR_LABEL"],
        value_vars=years,
        var_name="Year",
        value_name="Value"
    )
    long_df["Year"] = pd.to_numeric(long_df["Year"], errors="coerce")
    long_df["Value"] = pd.to_numeric(long_df["Value"], errors="coerce")
    return long_df

ict_long = to_long(ict_sa, ict_years)
cyb_long = to_long(cyb_sa, cyb_years)
print("\nICT Long format sample:")
print(ict_long.head(5))
print("\nCyber Long format sample:")
print(cyb_long.head(5))
```

```python
# Pivot to Year × Indicator
ict_pivot = ict_long.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")
cyb_pivot = cyb_long.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")
print("\nICT Pivot (Year × Indicators):", ict_pivot.shape)
print(ict_pivot.head(3))
print("\nCyber Pivot (Year × Indicators):", cyb_pivot.shape)
print(cyb_pivot.head(3))
```

```python
# YoY analysis
arr = ict_pivot.to_numpy(dtype=float)
ict_diff = np.diff(arr, axis=0)
ict_pct  = np.divide(ict_diff, arr[:-1], out=np.full_like(ict_diff, np.nan), where=arr[:-1]!=0)
print("\nICT YoY Absolute Change sample:")
print(pd.DataFrame(ict_diff, index=ict_pivot.index[1:], columns=ict_pivot.columns).head(3))
print("\nICT YoY % Change sample:")
print(pd.DataFrame(ict_pct, index=ict_pivot.index[1:], columns=ict_pivot.columns).head(3))
```

```python
# Composite indices
ict_comp = np.nanmean(arr, axis=1)
print("\nICT Composite sample (first 5 years):")
print(dict(zip(ict_pivot.index[:5], ict_comp[:5])))
```

```python
# CAGR per indicator
def cagr_per_indicator(df_long):
    rows = []
    for ind, g in df_long.sort_values("Year").groupby("INDICATOR_LABEL"):
        g = g.dropna(subset=["Value"])
        if g.empty: continue
        first_val, last_val = g.iloc[0]["Value"], g.iloc[-1]["Value"]
        n_years = int(g.iloc[-1]["Year"] - g.iloc[0]["Year"])
        if n_years > 0 and first_val not in [0, np.nan]:
            cagr = (last_val / first_val) ** (1/n_years) - 1
            rows.append((ind, cagr))
    return pd.DataFrame(rows, columns=["Indicator","CAGR"]).sort_values("CAGR", ascending=False)

ict_cagr = cagr_per_indicator(ict_long)
print("\nICT CAGR sample:")
print(ict_cagr.head(5))
```

```python
TOP_N = 12          # number of indicators to display
LABEL_MAX = 42      # max characters for labels

df = ict_pivot.copy()

# Keep years with any data; find global start/end years with broad coverage
valid_years = df.dropna(how='all').index
start_year = int(valid_years.min())
end_year   = int(valid_years.max())

# Rank indicators: prioritize completeness across the full window, then variance
comp_full = df.loc[[start_year, end_year]].notna().sum(axis=0)  # need data in both years
variance  = pd.Series(np.nanvar(df.to_numpy(float), axis=0), index=df.columns)

cands = pd.DataFrame({"comp_full": comp_full, "var": variance}).dropna()
ranked = cands.sort_values(["comp_full", "var"], ascending=[False, False]).index.tolist()

# Select top-N indicators that have both start and end values
cols = []
for c in ranked:
    if len(cols) >= TOP_N: break
    if pd.notna(df.at[start_year, c]) and pd.notna(df.at[end_year, c]):
        cols.append(c)

if not cols:
    raise ValueError("No indicators have values at both the start and end year.")

# Build start/end series
start = df.loc[start_year, cols]
end   = df.loc[end_year, cols]

# Sort by absolute change for a clean story
order = (end - start).abs().sort_values(ascending=True).index  # small→large slope for readability
start = start[order]
end   = end[order]

# Plot
y = np.arange(len(order))
fig_h = max(6, 0.45 * len(order))
fig, ax = plt.subplots(figsize=(12, fig_h))

for i, ind in enumerate(order):
    ax.plot([0, 1], [start[i], end[i]], marker='o')
    ax.text(-0.02, start[i], f"{start[i]:.2f}", va='center', ha='right', fontsize=8)
    ax.text( 1.02, end[i],   f"{end[i]:.2f}",   va='center', ha='left',  fontsize=8)

# Y-axis labeling with indicator names in the middle (between start and end)
mid_y = (start.values + end.values) / 2
for i, ind in enumerate(order):
    ax.text(0.5, mid_y[i], shorten(str(ind), width=LABEL_MAX, placeholder="…"),
            ha='center', va='center', fontsize=9)

ax.set_xlim(-0.15, 1.15)
ax.set_xticks([0, 1])
ax.set_xticklabels([str(start_year), str(end_year)])
ax.set_title(f"ICT — Start vs End Levels (Slopegraph, {start_year} → {end_year})")
ax.set_ylabel("Level")
ax.grid(axis='y', alpha=0.2)

plt.tight_layout()
plt.show()
```

Insights

The start-vs-end slopegraph for South Africa’s ICT indicators shows large positive deltas for capacity and adoption metrics—most notably lit/equipped international bandwidth, internet traffic (mobile and fixed), fixed-broadband subscriptions, and 5G population coverage. Foundational coverage (2G/3G/4G) is largely flat, signalling saturation at those layers. Price baskets (voice/data and fixed 5GB) move downward over the window, indicating improved affordability. Legacy fixed-telephone lines exhibit a clear structural decline, while active mobile-broadband subscriptions are comparatively stable to softer, even as usage intensity rises. Taken together, the end-state levels are materially higher for data-centric indicators than at the start.

Conclusions

The slopegraph evidences a structural shift from voice to data, with the system moving from basic coverage build-out to higher capacity and quality (throughput, 5G) and greater usage intensity. Affordability gains (falling price indices) align with the lift in adoption and traffic, while mature layers show limited headroom and legacy services continue to contract. Net-net, the sector’s terminal position is stronger than its baseline, confirming a long-run upward trend with modernization concentrated in high-speed access, capacity, and data consumption.

```python
# --- Helper: z-score normalisation within each indicator (column) ---
def zscore_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    # compute mean/std per column ignoring NaNs
    col_mean = df.mean(axis=0, skipna=True)
    col_std  = df.std(axis=0, skipna=True)
    # avoid division by zero
    col_std = col_std.replace(0, np.nan)
    return (df - col_mean) / col_std

# Normalise each pillar so indicators are comparable before averaging
ict_norm = zscore_df(ict_pivot)
cyb_norm = zscore_df(cyb_pivot)

# Composite = NaN-aware mean across indicators per year
ict_comp = ict_norm.mean(axis=1, skipna=True)
cyb_comp = cyb_norm.mean(axis=1, skipna=True)

# Align to common South Africa years where at least one pillar has a value
years_common = sorted(set(ict_comp.dropna().index).intersection(set(cyb_comp.dropna().index)))
if not years_common:
    raise ValueError("No common years with composite values in both ICT and Cyber.")

ict_line = ict_comp.loc[years_common]
cyb_line = cyb_comp.loc[years_common]

# Plot
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ict_line.index.astype(int), ict_line.values, marker='o', label="ICT (normalized composite)")
ax.plot(cyb_line.index.astype(int), cyb_line.values, marker='o', label="Cyber (normalized composite)")

ax.set_title("ICT vs Cyber — Normalized Composite (Common Years, South Africa)")
ax.set_xlabel("Year")
ax.set_ylabel("Composite (z-score mean)")
ax.legend()
plt.tight_layout()
plt.show()
```

Insights

On the common-year window, the normalized composites place ICT and Cyber on the same scale (z-scores): values above 0 indicate performance above each pillar’s long-run average and values below 0 indicate below average. The ICT line sits consistently above 0 and above the Cyber line, signalling relative outperformance and sustained positive momentum. Cyber starts below its average and rises into positive territory by the end of the window, indicating a catch-up from a lower base. The spread between the two pillars narrows, but ICT maintains the lead. Importantly, these indicators reflect system capability and maturity, not public awareness—so the data support a maturity gap (ICT adoption/capacity ahead of Cyber capability) rather than any claim about what people “knew”.

Conclusion

On a like-for-like, normalized basis, ICT outperforms Cyber across the period, maintaining an above-average trajectory while Cyber improves from below-average to above-average, narrowing (but not closing) the gap. This pattern suggests South Africa has scaled access, capacity, and usage faster than it has strengthened cybersecurity readiness, i.e., adoption is outpacing protection.ICT’s structural position remains stronger; Cyber shows meaningful progress from a lower starting point.

```python
KEYWORDS = ["price", "basket", "voice", "data-only", "usage", "bundle", "tariff", "5gb"]
TOP_N = 6            # max number of price indicators to plot
LABEL_MAX = 50       # max characters for legend labels

df = ict_long.copy()

# Ensure numeric
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["Value"] = pd.to_numeric(df["Value"], errors="coerce")

# Filter to likely price/affordability indicators by label keywords (case-insensitive)
mask = df["INDICATOR_LABEL"].str.lower().fillna("").apply(
    lambda s: any(k in s for k in KEYWORDS)
)
price_df = df[mask].copy()

if price_df.empty:
    raise ValueError("No pricing/affordability indicators found. Adjust KEYWORDS to match your labels.")

# Pivot to Year × Indicator for processing
price_pivot = price_df.pivot_table(index="Year", columns="INDICATOR_LABEL", values="Value", aggfunc="mean")

# Rank indicators by completeness, then variance (most informative first)
completeness = price_pivot.notna().sum(axis=0)
variance = pd.Series(np.nanvar(price_pivot.to_numpy(float), axis=0), index=price_pivot.columns)
ranked = (pd.DataFrame({"comp": completeness, "var": variance})
          .sort_values(["comp", "var"], ascending=[False, False])).index.tolist()

# Keep top-N
cols = ranked[:TOP_N]
price_pivot = price_pivot[cols]

# Determine common years across selected indicators and choose base year
valid_years_per_col = [set(price_pivot[c].dropna().index.astype(int)) for c in price_pivot.columns]
common_years = sorted(set.intersection(*valid_years_per_col)) if valid_years_per_col else []
if len(common_years) < 2:
    raise ValueError("Not enough common years across selected price indicators to index. Try increasing TOP_N or adjusting KEYWORDS.")
base_year = common_years[0]

# Build indexed series (Base Year = 100) and drop any series with invalid base values
indexed = {}
for c in price_pivot.columns:
    s = price_pivot[c].dropna()
    if base_year not in s.index or s.loc[base_year] in [0, np.nan]:
        continue
    indexed[c] = (s.loc[common_years] / s.loc[base_year]) * 100.0

if not indexed:
    raise ValueError("No valid series could be indexed (base year missing or zero).")

# Plot
fig, ax = plt.subplots(figsize=(12, 5))
for label, series in indexed.items():
    ax.plot(series.index.astype(int), series.values, marker='o', label=shorten(str(label), width=LABEL_MAX, placeholder="…"))

ax.set_title(f"ICT Prices/Affordability — Indexed to {base_year} = 100 (South Africa)")
ax.set_xlabel("Year")
ax.set_ylabel("Index (Base Year = 100)")

# Reference line at 100 (base)
ax.axhline(100, linestyle='--', linewidth=1)

# Place legend outside for readability
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), fontsize=9)
plt.tight_layout()
plt.show()
```

```python
TOP_N = 10       # fastest growers to show
BOTTOM_N = 5     # biggest decliners to show
LABEL_MAX = 48   # max characters for indicator labels in the plot


df = cyb_long.copy()

# Ensure numeric
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["Value"] = pd.to_numeric(df["Value"], errors="coerce")

# Compute CAGR per indicator (first -> last non-NaN)
cagr = {}
for name, g in df.groupby("INDICATOR_LABEL"):
    g = g.sort_values("Year")
    g = g[pd.notna(g["Value"])]
    if g.shape[0] < 2:
        continue
    y0, v0 = g.iloc[0][["Year","Value"]]
    y1, v1 = g.iloc[-1][["Year","Value"]]
    years = int(y1 - y0)
    # CAGR requires positive start/end and at least 1 year span
    if years <= 0 or pd.isna(v0) or pd.isna(v1) or v0 <= 0 or v1 <= 0:
        continue
    cagr[name] = (v1 / v0) ** (1/years) - 1

cagr = pd.Series(cagr).dropna()
if cagr.empty:
    raise ValueError("No valid CAGR could be computed from cyb_long.")

# Select top/bottom movers and order for display
sel = pd.concat([cagr.sort_values(ascending=False).head(TOP_N),
                 cagr.sort_values(ascending=True).head(BOTTOM_N)])
sel = sel.sort_values()  # ascending for clean vertical ordering

# Convert to % for readability
sel_pct = sel * 100.0
labels = [shorten(k, width=LABEL_MAX, placeholder="…") for k in sel_pct.index]

# Plot (lollipop)
y = np.arange(len(sel_pct))
fig_h = max(6, 0.45 * len(sel_pct))
fig, ax = plt.subplots(figsize=(12, fig_h))

ax.hlines(y, 0, sel_pct.values)     # sticks
ax.plot(sel_pct.values, y, 'o')     # lollipops

ax.set_yticks(y)
ax.set_yticklabels(labels, fontsize=9)
ax.set_xlabel("CAGR (% per year)")
ax.set_title("Cyber — CAGR Ranking (Top & Bottom Movers)")

# Zero reference line
ax.axvline(0, linestyle='--', linewidth=1)

# Annotate values
for yi, val in zip(y, sel_pct.values):
    ax.text(val + (0.5 if val >= 0 else -0.5), yi, f"{val:.1f}%", va='center',
            ha='left' if val >= 0 else 'right', fontsize=8)

plt.tight_layout()
plt.show()
```
